_**(a) La mayor parte de los lenguajes de programación y muchos programas con orientación científica incluyen generadores  de  números  pseudoaleatorios  con  distribución  uniforme  en  [0,1).  Cualquier  conjuto  de dichos números constituye una muestra de una variable aleatoria continua (tema de la guía siguiente) con dicha distribución, pero aquí nos interesa solamente el hecho de que pueden ser usados para simular experimentos de Bernoulli.  Mostrar que la probabilidad de que tal generador entregue un número en [0,p) (0≤p≤1) es p(y en [p,1) es 1−p).  Implemente, a partir de esta propiedad, un programa que genere el resultado de una serie de n experimentos de Bernoulli independientes y cuente el número de éxitos obtenido.
````python
%matplotlib inline 

from __future__ import division

import numpy as np
import matplotlib.pyplot as plt
````

Se utilizó un generador de números pseudo-aleatorios uniformes en [0,1] para generar una variable aleatoria con distribución binomial $B(k|n,p)$:

````python
def gen_binomial(n,p):
    '''
    Genera una v.a. que cuenta la cantidad de éxitos de n experimentos de Bernoulli
    con una probabilidad de éxito p. La variable exitos ~ Binomial(n,p)
    '''
    # Se generan N puntos con distribución U[0,1]
    x = np.random.uniform(0,1,np.int_(n))
    # Se cuentan la cantidad de éxitos
    exitos = np.size(np.where(x<=p))
    return exitos
````

Se definió otra función para repetir el experimento anterior `nrep` veces:

````python
def array_binomial(n,p,nrep):
    '''
    Genera nrep realizaciones de una variable con distribución Binomial(n,p)
    '''
    array_exitos = []
    for j in range(1,nrep+1):
        array_exitos.append(gen_binomial(n,p))
    return np.asarray(array_exitos)
````

_**(b)  Suponga que inciden exactamente n=15 fotones sobre un detector con una eficiencia $\epsilon = 0.75$.  Use el programa escrito en el punto anterior para determinar cuántos fotones son detectados en un experimento particular. Repita el experimento 1000 veces y realice un histograma de los resultados (i.e., del número de fotones detectados en cada experimento). Compárelo con la distribución de probabilidad teórica para dicha variable aleatoria. No olvide normalizar correctamente el histograma para realizar la comparación.**_

Se escribió una función auxiliar con el objetivo de normalizar y hacer cálculos secundarios que serán útiles para hacer los gráficos con histogramas:

````python
def mi_histograma(x):
    '''
    Función que calcula las entradas y los bines en un histograma.
    Pensado para distribuciones discretas, con una clase para cada
    posible valor de la v.a.
    
    Entrada:
        x: Array con las realizaciones de una dada variable aleatoria
    Salida:
        k_nor: entradas normalizadas en cada clase
        sig_nor: incertezas (normalizadas) para cada clase.
                 (Asumiendo que el número de entradas es Poisson)
        bin_center: array con la posición del centro de cada clase
        bin_left: array con la posición del lado izquierdo de cada clase
        '''
    # Fijo las posiciones del extremo izquierdo de los bines
    mis_bins = np.arange(min(x),max(x)+2,1)

    # En mis_bins y bin_left son iguales en este caso
    k, bin_left = np.histogram(x,bins=mis_bins)
    # Ancho de cada bin
    dbin = bin_left[1]-bin_left[0]
    # Vector con la posición central de cada bin
    bin_center = 0.5*(bin_left[1:]+bin_left[:-1]) - dbin

    # Normalización de las cuentas detectadas
    k_nor = k/dbin/np.sum(k)
    # Se asume incertezas poissonianas y se las normaliza
    sig_nor = k**0.5/dbin/sum(k)
    
    return k_nor, sig_nor, bin_center, bin_left
````

Se generan 1000 repeticiones de la simulación consistente en obtener la cantidad de fotones detectados cuando llegan $n=15$ a un detector con eficiencia $\epsilon=0.75$

````python
# Cantidad de fotones que llegan al detector
n = 15
# Eficiencia  del detector
epsilon = 0.75
# Repeticiones del experimento
nrep1 = 1000
# Simulación del experimento
t = array_binomial(n,epsilon,nrep1)

k_nor, k_sig_nor, bin_center, bin_left = mi_histograma(t)

# Gráfico del histograma
fig1, ax1 = plt.subplots()
ax1.bar(
    bin_center,
    k_nor,
    color = (0.8,0.8,0.8),
    width = bin_center[1]-bin_center[0],
    yerr = k_sig_nor,
    error_kw=dict(ecolor=(0.3,0.3,0.3), lw=2, capsize=2, capthick=2),
    label='Histograma simulado'
)

ax1.set_xlabel('Fotones detectados, k')
ax1.set_ylabel('Probabilidad')
ax1.set_xticks(bin_left[:-1]);
ax1.set_title('Llegan {} fotnes a un detector con $\epsilon$ = {}'.format(n,epsilon))

ax1.legend(numpoints=1,loc='best')
plt.grid(True)
plt.show()
````

**Comparación con la distribución teórica**

Se grafica la distribución teórica correspondiente a una binomial B(k|n=15 p=0.75) siendo:

\begin{equation}
B(k|n,p) = {n \choose k}p^k(1-p)^{n-k}
\end{equation}

### Simulación de un proceso de Poisson

_**(c) Ahora considere una fuente de intensidad media I=15 fot $s^{−1}$. Simule el número de fotones emitidos por la fuente en ∆t=1 s del siguiente modo.  Subdivida el intervalo ∆t en m >> 1 subintervalos iguales dt (sugerencia: use m=1000).  Aproxime la probabilidad de que la fuente emita un fotón en dt como Idt , y desprecie la probabilidad de emitir más de 1 fotón en el mismo intervalo (¿cómo justifica estas hipótesis?). Simule entonces el número de fotones emitidos en cada dt usando el programa desarrollado en el punto (a)  (¿por  qué es válido usar un programa que simula experimentos de Bernoulli?), y sumando sobre todos los dt calcule el número total de fotones emitidos durante ∆t. Repita el experimento 1000 veces, construya un histograma de los resultados y superponga a éste la distribución teórica correspondiente (no olvide normalizar). Discuta el procedimiento y los resultados en base a las hipótesis del proceso de Poisson.**_


En este punto se simulará la emisión de fotones de una fuente. La fuente emite en promedio $I=15$ fotones por segundo, y se busca encontrar la distribución de la cantidad de fotones emitidos por segundo. Para ello se divide el tiempo en 1000 intervalos ( $dt=0,001$ s ), y se considera que Idt es la probabilidad de que emita en ese $dt$. Se supondrá que no pueden haber más de una emisión en ese intervalo. Esta hipótesis se justi






ca ya que se están mirando intervalos temporales de 0,001 s , y en promedio, uno esperaría encontrar 0,015 fotones (cantidad despreciable). Se simuló, entonces, una distribución $B(k|1,Idt)$ en cada intervalo, y luego se sumó la cantidad de fotones para encontrar la emisión en 1 s.















Se definen las magnitudes a utilizar:

````python
# Tiempo total [s]
Dt = 1
# Intensidad de la fuente [1/s]
I = 15
# Número de puntos de discretización temporal
m = 1000
# Diferencial temporal
dt = Dt/m
# Probabilidad de emitir un fotón en un dt
p = I*dt
# Repeticiones del experimento
nrep2 = 1000
````

Se realiza la simulación con los gráficos correspondientes:

````python
# Cuenta la cantidad de emisiones en un dado dt, para cada experimento (nrep2)
emi_fuente_dt = np.zeros(nrep2)
# Loop sobre cada dt donde se acumulan los fotones emitidos
# Se repite este experimento nrep2 veces 
for i in range(1,m+1):
    emi_fuente_dt += array_binomial(1,p,nrep2)

# Emisiones totales
emi_fuente = emi_fuente_dt

# Calculo las cuentas normalizadas, con su incerteza y los bines centrados
k_nor, k_sig_nor, bin_center, bin_left = mi_histograma(emi_fuente)

# Gráfico del histograma
fig2, ax2 = plt.subplots()
ax2.bar(
    bin_center,
    k_nor,
    color = (0.8,0.8,0.8),
    width = bin_center[1]-bin_center[0],
    yerr = k_sig_nor,
    error_kw=dict(ecolor=(0.3,0.3,0.3), lw=2, capsize=2, capthick=2),
    label='Histograma simulado'
)

ax2.set_xlabel('Fotones emitidos por segundo, r')
ax2.set_ylabel('Probabilidad')
ax2.set_ylim(0,0.165)
ax2.set_title('Fuente que emite con una intensidad de {} fotones/s'.format(I))

# Se compara con la distribución teóricas
from scipy.stats import poisson

markline=ax2.stem(
    bin_left[:-1],
    poisson.pmf(bin_left[:-1],I),
    'r',
    label='P(r|$\mu$={})'.format(I))

plt.setp(markline,'markerfacecolor','r')

ax2.legend(numpoints=1,loc='best')
plt.grid(True)
plt.show()
````

**Distribución teórica**

En este caso los fotones emitidos por segundo sigue una distribución teórica de Poisson con parámetro $\mu=I$:

\begin{equation}
P(r | \mu) = e^{-\mu}\frac{\mu^r}{r!}
\end{equation}

### Simulación conjunta emisión-detección

_**(d) Dado el resultado de cada uno de los 1000 experimentos del punto anterior, use el mismo programa para calcular el número de fotones detectados por el detector del punto (b), suponiendo que éste opera durante ∆t = 1 s, y que todos los fotones emitidos llegan a él (i.e., que el detector subtiende un ángulo sólido de 4π visto desde la fuente). Realice el histograma correspondiente y compárelo con la distribución teorica (no olvide normalizar). Discuta los resultados (vea el ejercicio 19b).**_

Una vez obtenida la cantidad de fotones emitida por la fuente (```emi_fuente```), se procedió a utilizar cada uno de los 1000 valores obtenidos para simular cuántos de todos ellos serían contados por el detector:

````python
fot_det = []
for r in np.nditer(emi_fuente):
    fot_det.append(gen_binomial(r,epsilon))
````

Es decir, para cada valor $r$ de fotones emitidos por segundo, se calculó $B(k|r,\epsilon)$ utilizando la función ```gen_binomial``` definida previamente. Se supuso que todos los fotones emitidos por la fuente llegaban al detector (detector de 4π ). Una vez obtenidos los valores de fotones detectados, se procedió a construir el histograma correspondiente:

````python
det_nor, det_sig_nor, bin_center, bin_left = mi_histograma(np.asarray(fot_det))

# Gráfico del histograma
fig3, ax3 = plt.subplots()
ax3.bar(
    bin_center,
    det_nor,
    color = (0.8,0.8,0.8),
    width = bin_center[1]-bin_center[0],
    yerr = det_sig_nor,
    error_kw=dict(ecolor=(0.3,0.3,0.3), lw=2, capsize=2, capthick=2),
    label='Histograma simulado'
)

ax3.set_xlabel('Fotones detectados por segundo, k')
ax3.set_ylabel('Probabilidad')
ax3.set_ylim(0,0.165)
ax3.set_title('Fuente de {} fotones/s en presencia de un detector con $\epsilon$ = {}'.format(I,epsilon))

# Comparación con la distribución teórica
markline=ax3.stem(
    bin_left[:-1],
    poisson.pmf(bin_left[:-1],I*epsilon),
    'r',
    label='P(k|$\mu$=$\epsilon$I)'
    )

plt.setp(markline,'markerfacecolor','r')

ax3.legend(numpoints=1,loc='best')
plt.grid(True)
plt.show()
````

La distribución teórica utilizada para el punto anterior corresponde a una de Poisson con parámetro $\mu=I\epsilon$. Esto se demuestra planteando cuál es la probabilidad $P(k)$ de que se detecten $k$ fotones por segundo:

\begin{equation}
P(k) = \sum_{r=0}^{r=+\infty} P(r|I) B(k|r,\epsilon)
\end{equation}

donde $P(n|I)$ es la distribución de Poisson para la cantidad de emisiones por segundo $n$; y $B(k|n,\epsilon)$ son la cantidad de detecciones que se producen en el detector con eficiencia $\epsilon$ cuando le llegan $n$ fotones.

Luego de un poco de álgebra, la expresión anterior se reduce a:

\begin{equation}
P(k) = P(k|I\epsilon)
\end{equation}

Es decir, que la cantidad de fotones detectados provenientes de la fuente seguirá una distribución de Poisson con parámetro $\mu=I\epsilon$.

Mirando una determinada clase de un histograma (y asumiendo un número fijo de eventos totales), se puede pensar que existen sólo
dos resultados posibles: que un valor caiga en esa clase o que no caiga. Por tal razón, el
número de cuentas en una clase tiene una distribución binomial.

Si en cambio, la cantidad total de entradas del histograma sigue una distribución de Poisson, entonces se puede demostrar que la cantidad de entradas en una determinada clase seguirá extrictamente una distribución de Poisson.

Además, debido a que en un histograma el número total de casos que analiza siempre es muy grande (en este problema siempre fue 1000), se puede tomar el límite Poissoniano de la binomial. Por lo tanto, el error en cada clase puede aproximarse por la raíz cuadrada del valor medio de cuentas en dicha clase.
